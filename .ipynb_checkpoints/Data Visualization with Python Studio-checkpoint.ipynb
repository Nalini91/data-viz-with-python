```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

ratings_df = pd.read_csv("C:/Users/Buddi/Documents/GitHub/data-viz-with-python/books.csv", error_bad_lines=False)

# ratings_df['publication_date'].head()
all_years = []

for y in ratings_df['publication_date']:
    date = y.split("/")
    all_years.append(date[2])
ratings_df["years"] = all_years
ratings_df.head()
```

```python
# Number of books published per year

books_per_year= ratings_df["years"].value_counts().to_frame().reset_index()
books_per_year
```

```python
books_per_year.rename(columns={'years':'year_counts'}, inplace=True)
```

```python
books_per_year.rename(columns={'index':'years'},inplace=True)
```

```python
books_per_year.head()
```

```python
ratings_df= ratings_df.merge(books_per_year,on="years",how = "inner")
ratings_df.head()
```

```python
ratings_df_1 = ratings_df[['year_counts','years']].drop_duplicates().head(20)
```

```python
ratings_df_1 = ratings_df_1.reset_index(drop=True)
```

```python
ratings_df_1
```

```python
ratings_df_1.to_csv("ratings_df_1.csv")
```

```python
# Use your Python skills to create a visualization here.

import mplcyberpunk
style.use('cyberpunk')
 
a = ratings_df_1['years']
b = ratings_df_1['year_counts']
 
# Figure Size
fig = plt.figure(figsize =(40, 40))
plt.xticks(rotation = 60, color = 'white', fontsize = 32,fontweight = 'bold')
plt.yticks(color = 'white', fontsize = 32, fontweight = 'bold')
plt.title('Num of books/year',color = 'white', fontsize = 50, fontweight = 'bold')
plt.xlabel('years',color = 'white', fontsize = 40, fontweight = 'bold')
plt.ylabel('year_counts',color = 'white',fontsize = 40, fontweight = 'bold')
plt.savefig('year-counts.png')
plt.bar(a,b)
#plt.legend(loc='lower right')
plt.show()
```

```python

# Number of books published by each publisher

books_by_publisher= ratings_df["publisher"].value_counts().to_frame().reset_index()
books_by_publisher
```

```python
books_by_publisher.rename(columns={'publisher':'Books_count'}, inplace=True)
```

```python
books_by_publisher.rename(columns={'index':'publisher'}, inplace=True)
```

```python
books_by_publisher
```

```python
ratings_df= ratings_df.merge(books_by_publisher,on="publisher",how = "inner")
ratings_df.head()
```

```python
ratings_df_1 = ratings_df[['Books_count','publisher']].drop_duplicates()
```

```python
ratings_df_1 = ratings_df_1.reset_index(drop=True).head(21)
```

```python
ratings_df_1
```

```python
ratings_df_1.to_csv("ratings_df_1.csv")
```

```python
# Use your Python skills to create a visualization here.

import mplcyberpunk
style.use('cyberpunk')
 
a = ratings_df_1['publisher']
b = ratings_df_1['Books_count']
 
# Figure Size
fig = plt.figure(figsize =(40, 10))
plt.xticks(rotation = 60, color = 'white', fontsize = 32,fontweight = 'bold')
plt.yticks(color = 'white', fontsize = 32, fontweight = 'bold')
plt.title('Num of books/publisher',color = 'white', fontsize = 50, fontweight = 'bold')
plt.xlabel('publisher',color = 'white', fontsize = 40, fontweight = 'bold')
plt.ylabel('Books_count',color = 'white',fontsize = 40, fontweight = 'bold')
plt.savefig('Books-counts.png')
plt.bar(a,b)
plt.show()
```
